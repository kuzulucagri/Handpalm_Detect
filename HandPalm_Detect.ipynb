```python
#We import our openCV and numpy library
import cv2
import numpy as np

#We have calc_line function for calculate line length of our images
def calc_line(img,img1):
    pixel_length = 0 #Declare 0 for line length
    pixel_array = [] #Declare null list for assign pixel_length to pixel_array
    lined = np.copy(img) * 0 #I don't wanna work on original image so I use that function
    #HoughLinesP is a transform used to detect straight lines.
    lines = cv2.HoughLinesP(img1, 1, np.pi / 180, 15, np.array([]), 50, 20) 
    for line in lines: #Our loop
        for x1, y1, x2, y2 in line:
            cv2.line(lined, (x1, y1), (x2, y2), (0, 0, 255)) #That is used for draw a line into this (x1,y1,x2,y2) parameters
    
            pixel_length = np.abs(x2 - x1) #We need to use np library for absolute calculation
            pixel_array.append(pixel_length) #We added our result into the list
        
            print("Pixel Length: {}".format(pixel_length)) #Print the screen our datas in every loop
    

    print("Average Pixel Length: {:.0f} pixel".format(np.average(pixel_array))) #That is equal to average of pixel_array

img = cv2.imread("palmHand.jpeg") #Our image path in here and it is into our same file 
original = cv2.resize(img,(800,600)) #We resize our image for reach maximum accuracy rate
cv2.imshow("original",original) #Display in screen our original image
img = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY) #We convert our image to gray scale 
cv2.imshow("gray",img) #Display in screen our gray scaled image

img = cv2.equalizeHist(img) #We use histogram equalization for improve contrast of our image

img = cv2.GaussianBlur(img, (9, 9), 0) #That function is about blurring data

img = cv2.Canny(img, 40, 80) #That function is about find edge side of image and in our palm project, canny is good way to reach better result.
cv2.imshow("my_image",img) # That is final result to image

w,h = img.shape #We declare width and height from our image with shape function
#I assigned properly algorithm for heart, head, life and fate lines
heart_line = img[250:w-220,150:h-300]
ori_heart_line = original[250:w-220,150:h-300]
head_line = img[290:w-200,220:h-300]
ori_head_line = img[290:w-200,220:h-300]
life_line = img[280:w-100,350:h-250]
ori_life_line = img[280:w-100,350:h-250]
fate_line = img[200:w-80,250:h-320]
ori_fate_line = img[200:w-80,250:h-320]

#We called function 
calc_line(ori_life_line,life_line)
calc_line(ori_heart_line,heart_line)
calc_line(ori_head_line,head_line)
calc_line(ori_fate_line,fate_line)    
    




cv2.waitKey(0) #If we click any pin on the keyboard then it will close the images
cv2.destroyAllWindows() #Close and clean all windows
```
