# We import library in here

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# We read our excel tables with pandas

```python
datas = pd.read_excel("palm_info.xlsx")
```

# Declare datas and gender

```python
result = datas[["life_line","head_line","heart_line","fate_line"]]
gender = datas.iloc[:,-1:].values
print(gender)
```

# We use label encoder for gender string to convert 0 and 1

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
gender[:,-1] = le.fit_transform(datas.iloc[:,-1])

print(gender)
```

# Train and test split function for our datas 

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(result,gender,test_size = 0.33, random_state = 0)
```

```python
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

# Scale data

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
#Scale for each other
X_train = sc.fit_transform(x_train)
X_test = sc.fit_transform(x_test)
```

```python
print(X_train)
print(X_test)
```

# Linear regression and predict

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(x_train,y_train)

y_predict = regressor.predict(x_test)

print(y_predict)
print(y_test)
```

```python
a = datas.iloc[:,0:1]
b = datas.iloc[:,3:4]
A = a.values
B = b.values

lin_reg = LinearRegression()
lin_reg.fit(A,B)

plt.scatter(A,B,color = "red")
plt.plot(a,lin_reg.predict(A),color = "blue")
plt.title("Linear Regression")
plt.show()

from sklearn.preprocessing import PolynomialFeatures
#2 degree
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(A)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,b)

plt.scatter(A,B,color = "red")
plt.plot(A,lin_reg2.predict(poly_reg.fit_transform(A)),color = "blue")
plt.title("Polynomial Regression with 2 degree")
plt.show()

#4 degree
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(A)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,b)

plt.scatter(A,B,color = "red")
plt.plot(A,lin_reg2.predict(poly_reg.fit_transform(A)),color = "blue")
plt.title("Polynomial Regression with 4 degree")
plt.show()
```
